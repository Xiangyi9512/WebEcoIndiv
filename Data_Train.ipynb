```python
train = pd.read_csv('..\\Coursework\\Dataset\\train.csv')
```

```python
adv_num = train['advertiser'].unique()
adv_num.sort()
train_stats = pd.DataFrame(index=adv_num, columns=['Bids', 'Imps', 'Clicks', 'Cost', 'CTR', 'CPM', 'eCPC'])
```

```python
def train_bids():
    for adv in adv_num:
        adv_rows = train[train['advertiser']==adv]
        adv_bids = adv_rows['bidprice'].sum()
        train_stats.loc[adv, 'Bids'] = adv_bids
```

```python
train_bids()
print(train_stats)
```

```python
def train_Imps():
    for adv in adv_num:
        adv_rows = train[train['advertiser']==adv]
        adv_Imps = adv_rows.shape[0]
        train_stats.loc[adv, 'Imps'] = adv_Imps
```

```python
train_Imps()
print(train_stats)
```

```python
def train_Imps():
    for adv in adv_num:
        adv_rows = train[train['advertiser']==adv]
        adv_Imps = adv_rows.shape[0]
        train_stats.loc[adv, 'Imps'] = adv_Imps
```

```python
def train_Clicks():
    for adv in adv_num:
        adv_rows = train[train['advertiser']==adv]
        adv_Clicks = adv_rows['click'].sum()
        train_stats.loc[adv, 'Clicks'] = adv_Clicks
```

```python
train_Clicks()
print(train_stats)
```

```python
def train_Cost():
    for adv in adv_num:
        adv_rows = train[train['advertiser']==adv]
        adv_Cost = (adv_rows['payprice']/1000).sum()
        train_stats.loc[adv, 'Cost'] = "{:.2f}".format(adv_Cost)
```

```python
train_Cost()
print(train_stats)
```

```python
def train_CTR():
    for adv in adv_num:
        adv_rows = train[train['advertiser']==adv]
        adv_Clicks = adv_rows['click'].sum()
        adv_Imps = adv_rows.shape[0]
        adv_CTR = adv_Clicks / adv_Imps * 100
        train_stats.loc[adv, 'CTR'] = "{:.2f}%".format(adv_CTR)
```

```python
train_CTR()
print(train_stats)
```

```python
def train_CPM():
    for adv in adv_num:
        adv_rows = train[train['advertiser']==adv]
        adv_CPM = adv_rows['payprice'].mean()
        train_stats.loc[adv, 'CPM'] = "{:.2f}".format(adv_CPM)
```

```python
train_CPM()
print(train_stats)
```

```python
def train_eCPC():
    for adv in adv_num:
        adv_rows = train[train['advertiser']==adv]
        adv_Cost = (adv_rows['payprice']/1000).sum()
        adv_Clicks = adv_rows['click'].sum()
        adv_eCPC = adv_Cost / adv_Clicks
        train_stats.loc[adv, 'eCPC'] = "{:.2f}".format(adv_eCPC)
```

```python
train_eCPC()
print(train_stats)
```
