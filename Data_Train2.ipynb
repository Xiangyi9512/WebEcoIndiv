```python
slotsize_price_info = train.groupby(['advertiser','slotwidth','slotheight'])['payprice'].mean()
slotsize_price_df = slotsize_price_info.reset_index()
slotsize_price_df
```

```python
from collections import OrderedDict
slotsize_price_slotwidth_index_1458 = slotsize_price_df.loc[(slotsize_price_df['advertiser'] == 1458)]['slotwidth'].values
slotsize_price_slotheight_index_1458 = slotsize_price_df.loc[(slotsize_price_df['advertiser'] == 1458)]['slotheight'].values
slotsize_price_slotwidth_index_3358 = slotsize_price_df.loc[(slotsize_price_df['advertiser'] == 3358)]['slotwidth'].values
slotsize_price_slotheight_index_3358 = slotsize_price_df.loc[(slotsize_price_df['advertiser'] == 3358)]['slotheight'].values
slotsize_price_1458 = slotsize_price_df.loc[(slotsize_price_df['advertiser'] == 1458)]['payprice'].values
slotsize_price_3358 = slotsize_price_df.loc[(slotsize_price_df['advertiser'] == 3358)]['payprice'].values
slotsize_price_index_1458  = [str(slotsize_price_slotwidth_index_1458[i]) + " * " + str(slotsize_price_slotheight_index_1458[i]) for i in range(len(slotsize_price_slotwidth_index_1458))]
slotsize_price_index_3358  = [str(slotsize_price_slotwidth_index_3358[i]) + " * " + str(slotsize_price_slotheight_index_3358[i]) for i in range(len(slotsize_price_slotheight_index_3358))]

slotsize_price_1458_dict = dict.fromkeys(slotsize_price_index_1458)
count_1458 = 0
for key, value in slotsize_price_1458_dict.items():
    slotsize_price_1458_dict[key] = slotsize_price_1458[count_1458]
    count_1458 += 1
slotsize_price_1458_dict = OrderedDict(sorted(slotsize_price_1458_dict.items(), key=lambda slotsize_price_1458_dict: slotsize_price_1458_dict[0]))

slotsize_price_3358_dict = dict.fromkeys(slotsize_price_index_3358)
count_3358 = 0
for key, value in slotsize_price_3358_dict.items():
    slotsize_price_3358_dict[key] = slotsize_price_3358[count_3358]
    count_3358 += 1

slotsize_price_3358_dict = OrderedDict(sorted(slotsize_price_3358_dict.items(), key=lambda slotsize_price_3358_dict: slotsize_price_3358_dict[0]))
```

```python
import numpy
def dict_compare(d1, d2):
    d1_keys = set(d1.keys())
    d2_keys = set(d2.keys())
    intersect_keys = d1_keys.intersection(d2_keys)
    added = d1_keys - d2_keys
    print(d2)
    for n in added:
        d2[n] = numpy.nan
    print(d2)
    return d2

slotsize_price_3358_dict = dict_compare(slotsize_price_1458_dict, slotsize_price_3358_dict)
slotsize_price_1458_dict = dict_compare(slotsize_price_3358_dict, slotsize_price_1458_dict)
```
